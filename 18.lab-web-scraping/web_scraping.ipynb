The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
RepoNames = []
for number in range(len(repo.find_all("h1",class_="h3 lh-condensed"))):
    name_repo = repo.find_all("h1",class_="h3 lh-condensed")[number].find("a").get("href")
    name_repo_clean = re.findall(r"\w+\/\w+",name_repo)
    RepoNames.append("".join(name_repo_clean))
````

#### Display all the image links from Walt Disney wikipedia page.

````python
# This is the url you will scrape in this exercise
url3 = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
# download the url

page3 = requests.get(url3)

#parse the html

Img = BeautifulSoup(page3.content, 'html.parser') 
````

````python
img_links = [Img.find_all("a" ,class_="image")[i].get("href") for i in range(len(Img.find_all("a" ,class_="image")))]
````

````python
print(*img_links, sep='\n')
````
