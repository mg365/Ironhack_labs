# Challenge 3 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```sklearn.cluster.KMeans(n_clusters=8, *, init='k-means++', n_init=10, max_iter=300, tol=0.0001, precompute_distances='deprecated', verbose=0, random_state=None, copy_x=True, n_jobs='deprecated', algorithm='auto')```
[info here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

````python
# Import library

from sklearn.cluster import KMeans
````

````python
# initialise the model

KM = KMeans(n_clusters=5, random_state = 123)

# fit the model

groceriesKM = KM.fit(groceriesScN)

# to add to the dataframe the labels assigned

groceries['labels_KM'] = groceriesKM.labels_
````

Count the values in `labels`.

````python
# The levels assigned to each:

groceriesKM.labels_
````

````python
# the unique value count from the labels

unique, counts = np.unique(groceriesKM.labels_, return_counts=True)

pd.DataFrame(np.asarray((unique, counts)).T, columns = ['value', 'count'])
````

````python
# the grouping after the clustering is done 

customers_agg=groceries.groupby("labels_KM").agg("mean")
customers_agg
````

````python
# k means determine k

from scipy.spatial.distance import cdist

distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(groceries)
    kmeanModel.fit(groceries)
    distortions.append(sum(np.min(cdist(groceries, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / groceries.shape[0])

    
# Plot the elbow

plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()

# so we can decide the optimal k (clusters) to chose from
# (in the previous model we used a k of 5)
````

# Challenge 4 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```sklearn.cluster.DBSCAN(eps=0.5, *, min_samples=5, metric='euclidean', metric_params=None, algorithm='auto', leaf_size=30, p=None, n_jobs=None)```
<br>
[documentation here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html)

````python
# import library

from sklearn.cluster import DBSCAN

# initialise model

DB = DBSCAN(eps=0.5)

# fit to data

groceriesDB = DB.fit(groceriesScN)

# obtain the labels and add to original DataFrame

groceries['labels_DBSCAN'] = groceriesDB.labels_
````

Count the values in `labels_DBSCAN`.

````python
# the unique value count from the labels

unique, counts = np.unique(groceriesDB.labels_, return_counts=True)

pd.DataFrame(np.asarray((unique, counts)).T, columns = ['valueDB', 'countDB'])
````
